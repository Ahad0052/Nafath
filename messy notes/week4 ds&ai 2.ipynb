```python
#columns = [age, heart_rate, cholesterol, blood_pressure]
import numpy as np
patients_data = np.array([
[45, 80, 200, 120],
[60, 90, 240, 140],
[30, 70, 180, 110],
[50, 85, 220, 130],
[65, 95, 260, 150]
])

#shape of dataset
print(patients_data.shape) #(5,4)
# how many patient 
print(patients_data.shape[0]) #5
# how many features
print(patients_data.shape[1]) #4
```

```python
#the data of the first patient
print(patients_data[0])
#all ages
print("\n",patients_data[:,:1])
#all blood pressure values
print("\n",patients_data[:,3:])
```

```python
#ave ages
totalages = np.sum(patients_data[:,:1]) 
numberages = patients_data.shape[1]
aveages = totalages / numberages
print("\n",aveages)
```

```python
#maximum heart rate
heartRate = patients_data[:,1:2]
print(np.max(heartRate))
```

```python
#minimum cholesterol leve
cholesterol = patients_data[:,2:3]
print(np.min(cholesterol))
```

```python
#blood pressure > 130
pressure = patients_data[:,3:]
print(pressure > 130)
```

```python
#cholesterol ≥ 240
cholesterol = patients_data[:,2:3]
print(cholesterol >= 240)
```

```python
#older than 50
age = patients_data[:,:1]
print(age >50)
```

```python
#high-risk patients
highP = pressure> 140
highC = cholesterol >240
print(highP * highC)
```

```python
total1 = 0
c1 = 0
total2 = 0
c2 = 0
ageHeart = patients_data[:,:2]
for i in range(patients_data.shape[0]):
    if ageHeart[:,0:1][i] >= 50:
        total1 += ageHeart[:,1:2][i]
        c1 += 1
        ave1 = total1 / c1
    else:
        total2 += ageHeart[:,1:2][i]
        c2 += 1
        ave2 = total2 / c2
        

print(ave1)
print(ave2)
if ave1 > ave2:
    print("above 50 have more heart rate")
else:
    print("under 50 have more heart rate")
```

```python
pressure = patients_data[:,3:]
print(pressure)
maxP = np.max(pressure)

heartRate = patients_data[:,1:2]
print(heartRate)
minH = np.min(heartRate)

for i in range(patients_data.shape[0]):
    if pressure[i] == maxP:
        print("patient at index ",i,"highst pressure (",maxP,")")
    if heartRate[i] == minH:
        print("patient at index ",i,"lowest heart rate (",minH,")")
```

```python
total1 = 0
c1 = 0
total2 = 0
c2 = 0
pressure = patients_data[:,3:]
age = patients_data[:,:1]
for i in range(patients_data.shape[0]):
    if age[i] >= 50:
        total1 += pressure[i]
        c1 += 1
        ave1 = total1 / c1
    else:
        total2 += pressure[i]
        c2 += 1
        ave2 = total2 / c2
        

print(ave1)
print(ave2)
if ave1 > ave2:
    print("older people have higher blood pressure")
else:
    print("younger people have higher blood pressuree")
```

```python
temp = np.array([30,0,73,0,67,7,2,0])

condition = temp != 0 
print(condition)
filtered = temp[condition]
print("non zeros",filtered)
avetemp = filtered.mean()
print("average",avetemp)
cleanedtemp = np.where(temp == 0, avetemp, temp)
print("original",temp)
print("cleaned",cleanedtemp)
```
