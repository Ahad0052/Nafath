```python
data = np.array([12,-20,1,4,-4,6,20,9,10])

q1 = np.quantile(data,0.25)
q2 = np.quantile(data,0.5)
q3 = np.quantile(data,0.75)
iqr = q3 - q1
minn = q1 - 1.5 * iqr
maxx = q3 + 1.5 * iqr
print("q1   ","| q2   ","|  q3","   | iqr","  |  min","    | max")
print(q1," | ",q2," | ",q3," | ",iqr," | ",minn," | ",maxx)
```

```python
#way 1
datafiltered = data[ (data < maxx)  & (data > minn ) ]
datafiltered
```

```python
out = data[ (data > maxx)  |  (data < minn ) ]
out
```

```python
filtered = np.where((data > minn) & (data < maxx),data,"out")
print(filtered)
```

```python
#way 2
out1 = np.where((data < minn) | (data > maxx), True, False)
print(out1)
# replace outliers with min or max depend on it
```

```python
data[out1]
```

```python
data = np.where((data < minn), datafiltered.min(), data)
data = np.where((data > maxx), datafiltered.max() , data )
print(data)
```
